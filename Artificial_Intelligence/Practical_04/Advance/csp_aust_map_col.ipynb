# 🌏 Practical 04 - Advanced CSP Australian Map Coloring

This notebook solves the classic **Australian Map Coloring Problem** using Constraint Satisfaction and backtracking graphs.

### 1. Definition and Setup

```python
import networkx as nx
import matplotlib.pyplot as plt
import time

neighbors = {
    'WA': ['NT', 'SA'],
    'NT': ['WA', 'SA', 'Q'],
    'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],
    'Q': ['NT', 'SA', 'NSW'],
    'NSW': ['Q', 'SA', 'V'],
    'V': ['SA', 'NSW'],
    'T': []
}

colors = ['red', 'green', 'blue', 'yellow']
assignment = {}

G = nx.Graph()
for state, nbrs in neighbors.items():
    for n in nbrs:
        G.add_edge(state, n)

pos = nx.spring_layout(G, seed=42)
```

### 2. Draw Helper Function

```python
def draw_graph():
    node_colors = [assignment.get(s, 'lightgray') for s in G.nodes()]
    plt.figure(figsize=(7,7))
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1000, font_size=14)
    plt.show()
    time.sleep(0.5)
```

### 3. Backtracking Solver with State Consistency Check

```python
def is_consistent(state, color):
    for neighbor in neighbors[state]:
        if neighbor in assignment and assignment[neighbor] == color:
            return False
    return True

def backtrack():
    if len(assignment) == len(neighbors):
        draw_graph()
        return assignment

    unassigned = [s for s in neighbors if s not in assignment][0]

    for color in colors:
        if is_consistent(unassigned, color):
            assignment[unassigned] = color
            draw_graph()
            result = backtrack()
            if result:
                return result
            del assignment[unassigned]
            draw_graph()
```

### 4. Run Backtracking

```python
solution = backtrack()
print("[SUCCESS] Final Assignment:", solution)
```

### 📌 Problem Statement

- Generate data following a normal distribution and plot a histogram.
- Compute mean, median, variance, and standard deviation.
- Simulate a coin toss experiment and compute the probability of heads.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Generate Normal Distribution Data

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
print(f"Generated {len(data)} samples from Normal Distribution")
```

```python
# Histogram Plot

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, color="steelblue", edgecolor="black")
plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Statistical Measures 

mean = np.mean(data)
median = np.median(data)
var = np.var(data)
std = np.std(data)

print(f"Mean                : {mean:.4f}")
print(f"Median              : {median:.4f}")
print(f"Variance            : {var:.4f}")
print(f"Standard Deviation  : {std:.4f}")
```

```python
# Coin Toss Exp and It's Probability

tosses = np.random.randint(0, 2, size=1000)

head_count = np.sum(tosses)
prob_heads = head_count / len(tosses)
tail_count = 1000 - head_count
prob_tails = 1 - prob_heads

print(f"Total Tosses : {len(tosses)}")
print(f"Heads Count  : {head_count}")
print(f"P(Heads)     : {prob_heads:.4f}")
print(f"Tails Count  : {tail_count}")
print(f"P(Tails)     : {prob_tails:.4f}")
```

### 📌 Problem Statement

- Create a 3×3 matrix and compute its determinant.
- Find the inverse of the matrix.
- Solve a system of linear equations:
  - 2x + 3y = 10
  - 4x + 5y = 20


```python
import numpy as np
```

```python
A = np.array([
    [2,3,1],
    [4,5,6],
    [7,8,9]
])
```

```python
print("Matix A: ")
print(A)
```

```python
det_A = np.linalg.det(A)
print(f"Determinant of arr: {det_A:.4f}")
```

```python
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("Inverse of A: ")
    print(np.round(inv_A, 4))

else:
    print("Matrix is Singular; inverse does not exist.")
```

```python
# Linear Equation: 2x + 3y = 10 & 4x + 5y = 20

coeff_matrix = np.array(
    [
        [2,3],
        [4,5]
    ]
)

const_vector = np.array(
    [10,20]
)

soln = np.linalg.solve(coeff_matrix, const_vector)
```

```python
print(f"Solution: x = {soln[0]:.4f}, y = {soln[1]:.4f}")
```

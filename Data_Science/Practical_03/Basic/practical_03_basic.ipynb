### 📌 Problem Statement

- Load the Iris dataset (UCI/sklearn).
- Perform data filtering based on conditions.
- Perform grouping by species.
- Compute aggregations: sum, mean, count, min, max.

```python
import pandas as pd 
```

```python
iris = pd.read_csv("Iris.csv")

print(iris.head())
print(f"Shape: {iris.shape}")
```

```python
# Data Filtering 

filtered_df = iris[iris['SepalLengthCm'] > 6.0]
print(f"Rows with Sepal Length > 6.0: {len(filtered_df)}")
print(filtered_df.head())
```

```python
# Grouping By Species 

grp = iris.groupby('Species')
print(f"Group Sizes: {grp.size()}")
```

```python
# Aggregation Function for Sepal Length

agg1 = grp["SepalLengthCm"].agg(['mean','sum','count','min','max'])

print("Aggregations For Sepal Length By Species: ")
print(agg1)
```

```python
# Aggregation Function for Sepal Width

agg2 = grp["SepalWidthCm"].agg(['mean','sum','count','min','max'])

print("Aggregations For Sepal Width By Species: ")
print(agg2)
```

```python
# Aggregation Function for Petal Length

agg3 = grp["PetalLengthCm"].agg(['mean','sum','count','min','max'])

print("Aggregations For Petal Length By Species: ")
print(agg3)
```

```python
# Aggregation Function for Petal Width 

agg4 = grp["PetalWidthCm"].agg(['mean','sum','count','min','max'])

print("Aggregations For Petal Width By Species: ")
print(agg4)
```
